```python
#Opening a Chrome browser
status = webdriver.Chrome(r'chromedriver.exe')

#Open the sign in page
status.get('https://www.saltybet.com/authenticate?signin=1')
  
#Letting the time to load the page
time.sleep(4)

#Signin in in both browsers
status.find_element_by_id(
                    "email").send_keys(email)
status.find_element_by_id(
                    "pword").send_keys(password)
#Submit the page (automatically goes to saltybet main page)
status.find_element_by_class_name('submit').click()

#Getting to the page whith the match status info
status.get('https://www.saltybet.com/shaker')
```

```python
#Defining simple functions that are just shorter forms of numerous repeted code
def stats_by_id(idt):
    return stats.find_element_by_id(idt).text

def salty_tag(idt, tag):
    return salty.find_element_by_id(idt).find_element_by_tag_name(tag).text
```

```python
# Handle operations performed in the while loop:
bet_status = "NULL"  # Handle control flow.
recorded = False #Tag if the match has been recorded or not
red, blue = "NULL", "NULL"  # Fighter.
match = 0 #Match counter
#Allocating an empty string value for all the future info variables
red_match_numb = blue_match_numb = red_odds = blue_odds = red_winrate = blue_winrate = red_tier = blue_tier = red_life = blue_life = red_meter = blue_meter = red_palette = blue_palette = red_author = blue_author = current_match_status = ""
stats.refresh()
status.refresh()


while True:
    try:
    #Using a try statement to prevent the loop for crashing everytime selenium can't get value during a transition
    #Retrieving the Betting status (4 possibilities : bets are open, bets are closed, there is a winner and a loser, the string is empty during the transitions)
        current_bet_status = salty.find_element_by_id(
                        "betstatus").text
        if len(current_bet_status) == 0:
                 pass
        elif bet_status == "":
                 bet_status = current_bet_status
        elif bet_status != current_bet_status:
            bet_status = current_bet_status


        # Get fighter names:
        red_current = salty_tag("sbettors1", "strong")
        blue_current = salty_tag("sbettors2", "strong")
        
        # Replacing if the fighters changed, and setting the "recorded" value to False:
        if red != red_current or blue != blue_current:
            recorded = False
            red = red_current
            blue = blue_current
            
        else:
            pass
    
        #Passing if there is an empty string while getting info from the stats broser (probably during refresh)
        if len(stats_by_id("p1totalmatches")) == 0 or stats_by_id("p2totalmatches") == 0:
            pass
        
        #If there is value in stats pages and variables have empty strings, it mean the loop has just been launched, or the match changed from the previous loop
        elif red_match_numb == "" or blue_match_numb == "":
            #Getting all he info from stats and status browsers
            #We do this BEFORE the match ends as they will disappear before the  match status is updated and the final loop begins

            red_match_numb = stats_by_id("p1totalmatches")
            blue_match_numb = stats_by_id("p2totalmatches")
            
            red_winrate = stats_by_id("p1winrate")
            blue_winrate = stats_by_id("p2winrate")
            
            red_tier = stats_by_id("p1tier")
            blue_tier = stats_by_id("p2tier")
            
            red_life = stats_by_id("p1life")
            blue_life = stats_by_id("p2life")
            
            red_meter = stats_by_id("p1meter")
            blue_meter = stats_by_id("p2meter")
            
            red_palette = stats_by_id("p1palette")
            blue_palette = stats_by_id("p2palette")
            
            red_author = stats_by_id("p1author")
            blue_author = stats_by_id("p2author")
        
            current_match_status = status.find_element_by_id(
                    "compendiumleft").text.split("\n")[4]
            
        if current_match_status[-16:] == 'next tournament!' or current_match_status == 'Tournament mode will be activated after the next match!':
            match_status = 'MatchMaking'
            
        elif current_match_status == 'Tournament mode start!' or current_match_status == 'FINAL ROUND! Stay tuned for exhibitions after the tournament!' or  current_match_status[-20:]== 'left in the bracket!':
            match_status = 'Tournament'
                
        elif current_match_status[-24:] == 'exhibition matches left!' or current_match_status == 'Matchmaking mode will be activated after the next exhibition match!':
            match_status = 'Exhib'
        
        else:
            pass
    

        #Similar thing with the odds, though these info change in their own pace 
        #(is empty when the fights end, so before the "win status" appears, and data 
        #only appears after the end of the betting time, so long after the other infos are filled), 
        #so they have to get their own if/else statements
        red_odds_current = salty.find_element_by_id(
        "odds").text.partition("$")[2].partition(" ")[0].replace(',', '')
        blue_odds_current = salty.find_element_by_id(
        "odds").text.partition("$")[2].partition(" ")[-1].partition("$")[-1].replace(',', '')
            
        if (len(red_odds_current) != 0 and len(blue_odds_current) != 0) and (red_odds != red_odds_current or blue_odds != blue_odds_current):              
            red_odds = red_odds_current
            blue_odds = blue_odds_current
            
        else:
            pass

        # 'win' status only happens after a match end, so the final loop begins if the match ended and hasn't been recorded
        if "win" in bet_status:
            if recorded is False:
            
                #Setting up the winner
                if red in bet_status:
                    print("Winner: {}, Loser: {}\n".format(red, blue))
                    winner = red
                
                elif blue in bet_status:
                    print("Winner: {}, Loser: {}\n".format(blue, red))
                    winner = blue
                
                #Writting the results
                with open('data.csv', 'a+', newline='') as write_obj:
                    # Putting values names in a list
                    dict_writer = DictWriter(write_obj, fieldnames=['status',
                                                                    'blue', 
                                                                    'blue_odds',  
                                                                    'blue_matches_number',
                                                                    'blue_winrate',
                                                                    'blue_tier',
                                                                    'blue_life',
                                                                    'blue_meter',
                                                                    'blue_palette',
                                                                    'blue_author',
                                                                    'red', 
                                                                    'red_odds', 
                                                                    'red_matches_number',
                                                                    'red_winrate',
                                                                    'red_tier',
                                                                    'red_life',
                                                                    'red_meter',
                                                                    'red_palette',
                                                                    'red_author',
                                                                    'winner'])
                    
                    # Creating a dictionary as a csv row with all the values
                    dict_writer.writerow({'status' : match_status,
                                            'blue' : blue, 
                                            'blue_odds' : blue_odds,  
                                            'blue_matches_number' : blue_match_numb,
                                            'blue_winrate' : blue_winrate,
                                            'blue_tier' : blue_tier,
                                            'blue_life' : blue_life,
                                            'blue_meter' : blue_meter,
                                            'blue_palette' : blue_palette,
                                            'blue_author' : blue_author,
                                            'red' : red, 
                                            'red_odds' : red_odds, 
                                            'red_matches_number' : red_match_numb,
                                            'red_winrate' : red_winrate,
                                            'red_tier' : red_tier,
                                            'red_life' : red_life,
                                            'red_meter' : red_meter,
                                            'red_palette' : red_palette,
                                            'red_author' : red_author,
                                            'winner' : winner})
                #Adding a match to our counter
                match += 1 
                print(
                        "{} matches watched.".format(match))
                #Setting recorded to True, preventing another match to be recorded
                recorded = True
                
                #Emptying all the variables
                red_match_numb = red_odds = blue_odds = blue_match_numb = red_winrate = blue_winrate = red_tier = blue_tier = red_life = blue_life = red_meter = blue_meter = red_palette = blue_palette = red_author = blue_author = current_match_status = ""
                
                #Refreshing pages that needs it
                stats.refresh()
                status.refresh()
                
                #Waiting 10 seconds so that the refreshing is finished
                time.sleep(10)
     
            else:
                pass

    except:
        pass
```
